```python
def classificaçãoSPI_(indice):
  if indice >=2:
    return "extremamente úmido"
  elif indice >=1.5 and indice <1.99:
    return "muito úmido"
  elif indice >=1 and indice <1.49:
    return "moderadamente úmido"
  elif indice <0.99 and indice >=-0.99:
    return "normal"
  elif indice >=-1.49 and indice <-1:
    return "moderadamente seco"
  elif indice >=-1.99 and indice <-1.5:
    return "severamente seco"
  else: 
    return "extremamente seco"

classificaçãoSPI_(-1.3025168940947887)
```

```python
# --------- Estatísticas

# ----- Média
Média=[]
Mínimo=[]
Máximo=[]
DesvioPadrão=[]
Mesorregioes=[]
Período=[]
ClassificaçãoSPI=[]

for j in range(len(data["listaAnos"].unique())):
  for i in range(len(data["MESO_"].unique())):
    Período.append(data["listaAnos"].unique()[j])
    Mesorregioes.append(data["MESO_"].unique()[i])
    Média.append(data[(data["MESO_"]==data["MESO_"].unique()[i]) & (data["listaAnos"]==data["listaAnos"].unique()[j])]["spi_12"].mean())
    Mínimo.append(data[(data["MESO_"]==data["MESO_"].unique()[i]) & (data["listaAnos"]==data["listaAnos"].unique()[j])]["spi_12"].min())
    Máximo.append(data[(data["MESO_"]==data["MESO_"].unique()[i]) & (data["listaAnos"]==data["listaAnos"].unique()[j])]["spi_12"].max())
    DesvioPadrão.append(data[(data["MESO_"]==data["MESO_"].unique()[i]) & (data["listaAnos"]==data["listaAnos"].unique()[j])]["spi_12"].std())
    ClassificaçãoSPI.append(classificaçãoSPI_(data[(data["MESO_"]==data["MESO_"].unique()[i]) & (data["listaAnos"]==data["listaAnos"].unique()[j])]["spi_12"].mean()))

tabelaAnaliseDescritiva=pd.DataFrame()
tabelaAnaliseDescritiva["Mesorregioes"]=Mesorregioes
tabelaAnaliseDescritiva["Período"]=Período
tabelaAnaliseDescritiva["Média"]=Média
tabelaAnaliseDescritiva["Mínimo"]=Mínimo
tabelaAnaliseDescritiva["Máximo"]=Máximo
tabelaAnaliseDescritiva["DesvioPadrão"]=DesvioPadrão
tabelaAnaliseDescritiva["ClassificaçãoSPI"]=ClassificaçãoSPI

tabelaAnaliseDescritiva
```

```python
def inserirLabels(i):
  if i==1:
    return "Sertão Pernambucano"
  
  elif i==2:
    return "São Francisco Pernambucano"
  
  elif i==3:
    return "Metropolitana do Recife"
  
  elif i==4:
    return "Agreste Pernambucano"
  
  else:
    return "Mata Pernambucana"
```

```python
# Gerando a base anual por meso

def filtrandoMesoEAno(df,meso):
  df_meso=df.loc[df['MESO_'] == meso]
  #df_ano=df_meso.loc[df_meso["listaAnos"] == ano]

  return df_meso

filtrandoMesoEAno(data,"Mata Pernambucana")
```

```python
#@title Texto de título padrão
def grafBoxPlot(df:dict, dadosEixoY:str,dadosEixoX:str, N: int,vetorCores:list,tituloEixoX:str,tituloEixoY:str):
  fig = go.Figure(data=[go.Box(
      y=filtrandoMesoEAno(df,df[dadosEixoX].unique()[i])[dadosEixoY],
      marker_color=vetorCores[i], 
      name = inserirLabels(i)
      ) for i in range(int(N))])

  # format the layout
  fig.update_layout( plot_bgcolor = 'white', showlegend=True,# title = meso+' (1998-2020)',
      xaxis=dict(showgrid=True, zeroline=False,gridwidth=1,gridcolor='lightgray', linewidth=1, showline=True,  \
                 showticklabels=True, title=tituloEixoX,),
      yaxis=dict(zeroline=False,gridcolor='lightgray',gridwidth=1, linewidth=1, showline=True, title=tituloEixoY),
      paper_bgcolor='rgb(255,255,255)', 
      font=dict( family="times new roman", size=20, color="black")
      #plot_bgcolor='rgb(233,233,233)',
  )
  return fig.show()

grafBoxPlot(data,"spi_12","MESO_",5,["#e41a1c","#377eb8", "#4daf4a", "#984ea3", "#ff7f00"],"Mesorregiões","SPI (12 meses)")
```
